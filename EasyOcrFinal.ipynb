```python
import easyocr
```

```python
reader = easyocr.Reader(['bn','en'], gpu=True)
```

```python
import os,re
directory = r'/content/drive/MyDrive/TestNidOld'
for filename in os.listdir(directory):
    current_file_name = directory+"/"+filename
    data = reader.readtext(current_file_name, detail=0)
    print(" =================================== ")
    print(filename)
    print(data)

    for item in reversed(data) :
        nid_match = re.search(r'\d{10,17}', item)
        if nid_match is None:
            line = re.sub('[^0-9.]', ' ', item)
            nid_search_data = line.replace(' ', '')
            nid_match = re.search(r'\d{10,17}', nid_search_data)
        if nid_match:
            nid_num = nid_match.group()
            print("NID : " + nid_num)

    for item in reversed(data) :
        dob_match = re.search(r'\d{2} [A-Za-z]{3} \d{4}', item)
        if dob_match:
            dob = dob_match.group()
            print("DOB : " + dob)

    print(" =================================== ")
    
```
